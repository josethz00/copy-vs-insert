```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
species = ("2 million", "5 million", "10 million")
commands_means = {
    'Parallel COPY': (5.910, 15.4, 21.19),
    'COPY': (10.008, 20.534, 36.965),
    'INSERT': (37.187, 78.957, 134.412),
}

x = np.arange(len(species))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in commands_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Seconds (s)')
ax.set_title('Performance in seconds by command and number of rows')
ax.set_xticks(x + width, species)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 150)

plt.show()
```
